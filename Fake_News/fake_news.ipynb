```python
import numpy as np
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from   sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
import nltk 
nltk.download('stopwords')
```

```python
print(stopwords.words('english'))
```

```python
df=pd.read_csv("train.csv")
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df=df.fillna('')
```

```python
#text data is very big to process in small project, merge author and title
```

```python
df['content']=df.author+df.title
df.head()
```

```python
x=df.drop(['label'],axis='columns')
y=df.label
```

```python
port_stem=PorterStemmer()
```

applied stemming and stopWord removal methods form IR

```python
def stemming(content):
    stem_content=re.sub('[^a-zA-z]',' ',content)
    stem_content=stem_content.lower()
    stem_content=stem_content.split()
    stem_content=[port_stem.stem(word) for word in stem_content if not word in stopwords.words('english')]
    stem_content=' '.join(stem_content)
    return stem_content
```

```python
df['content']=df['content'].apply(stemming)
```

```python
df.head()
```

```python
x=df.content.values
y=df.label.values
```

uses TF-IDF method to mark the numerical values

```python
vectorizer=TfidfVectorizer()
x=vectorizer.fit_transform(x)
```

```python
print(x[:6])
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.8)
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
x_new=x_test[0]
pred=model.predict(x_new)
if pred[0]==0:
    print("News is Real")
else:
    print("News is Fake")
```
